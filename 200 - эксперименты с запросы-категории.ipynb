```python
# df = dataAccess.getKeywordsLatestWeek()
# df['keyword'] = df['keyword'].str.strip()
# df = df[df['keyword'].str.contains('\d{6,}') == False]
df = df.groupby('keyword').sum().reset_index().sort_values('searches', ascending=False)

import pymorphy2
import numpy as np
from nltk.corpus import stopwords

morph = pymorphy2.MorphAnalyzer()

def normalize(word):
    if word == None: return word
    return morph.parse(word)[0].normal_form

stopWords = stopwords.words("russian")

def removeStopWords(word):
    if word == None: return word
    if word not in stopWords:
        return word
    else:
        return None

```

```python
df2 = pd.DataFrame(df['keyword'].str.split(' ', expand=True))

for column in df2.columns:
    # df2.loc[df2[column]=='кроссовки', column] = None
    # df2.loc[df2[column]=='женские', column] = None
    df2[column] = df2[column].apply(removeStopWords)
    df2[column] = df2[column].apply(normalize)

for index, row in df2.iterrows():
    df2.loc[index] = df2.loc[index].sort_values().tolist()

columnsSplited = df2.columns.to_list()

df2 = df2.fillna('')

df2['keyword'] = df['keyword']
df2['searches'] = df['searches']

df2 = df2.sort_values('searches', ascending=False)
```

```python
dfGroups = df2.groupby(columnsSplited)


dfAggregated = dfGroups.sum(numeric_only=True).sort_values('searches')

df3 = df2.drop_duplicates(subset=columnsSplited).set_index(columnsSplited)

df3 = df3[['keyword']].merge(dfAggregated, left_index=True, right_index=True).reset_index()[['keyword', 'searches']]

df3#.to_clipboard()
```
